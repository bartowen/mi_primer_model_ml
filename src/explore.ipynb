```python
# Eliminar información irrelevante
#  identificar a los clientes existentes que tienen una mayor probabilidad de contratar un depósito a largo plazo
# age. SI: la edad si es relevante, ya que es probable que a mayor edad sea más consciente de las inversiones y tengan en mente un deposito a plazo
# relevantes = age, job, marital, education, default, housing, loan, campaign, pdays, poutcome, emp.var.rate, cons.price.idx, cons.conf.idx, EURIBOR 3, y
# irrelevantes = contact, month, day_of_week, duration, previous, nr.employed
```

```python
columnas_eliminar=['contact', 'month', 'day_of_week', 'duration', 'previous', 'nr.employed']
df.drop(columnas_eliminar,axis=1,inplace=True)
df.info()
```

```python
# Analisis univariante
# Analisis Variables categoricas. Una variable categórica es un tipo de variable que puede tomar uno de un número limitado de categorías o grupos
# Hay que usar Histograma e identificar cuales son categoricas
# Variables Categoricas: job, marital, education, default, housing, loan, poutcome, y

import matplotlib.pyplot as plt
import seaborn as sns

fig, axis = plt.subplots(4, 2, figsize = (10, 15))

sns.histplot(ax= axis[0,0], data = df, x='job').tick_params(axis='x', rotation=45)
sns.histplot(ax= axis[0,1], data = df, x='marital').tick_params(axis='x', rotation=45)
sns.histplot(ax= axis[1,0], data = df, x='education').tick_params(axis='x', rotation=90)
sns.histplot(ax= axis[1,1], data = df, x='default').tick_params(axis='x', rotation=45)
sns.histplot(ax= axis[2,0], data = df, x='housing')
sns.histplot(ax= axis[2,1], data = df, x='loan')
sns.histplot(ax= axis[3,0], data = df, x='poutcome')
sns.histplot(ax= axis[3,1], data = df, x='y')

# Ajustar el layout
plt.tight_layout()

plt.show()
```

```python
# hay una presencia mayoritaria de clientes con estudios universitarios
# no tienen cliente vigente 
# un pco más de la mitad del dataset tiene un credito hipotecario, pero una gran participacion no tiene un prestamo personal (credito de consumo)
# El exito de la campaña anterior es muy bajo, ya que gran porcentaje esta identificado sin registro.
# menos de 5 mil han contratado DP
```

```python
# Analisis Univariente : numérica
# Variantes númericas: 'age', 'campaign', 'emp.var.rate', 'cons.price.idx', 'cons.conf.idx', 'euribor3m'

fig, axis = plt.subplots(6, 2, figsize = (10, 14), gridspec_kw={'height_ratios': [6, 1, 6, 1, 6, 1]})

# Crear una figura múltiple con histogramas y diagramas de caja

sns.histplot(ax = axis[0, 0], data = df, x = "age").set(xlabel = None)
sns.boxplot(ax = axis[1, 0], data = df, x = "age")

sns.histplot(ax = axis[0, 1], data = df, x = "campaign").set(xlabel = None, ylabel = None)
sns.boxplot(ax = axis[1, 1], data = df, x = "campaign")

sns.histplot(ax = axis[2, 0], data = df, x = "emp.var.rate").set(xlabel = None)
sns.boxplot(ax = axis[3, 0], data = df, x = "emp.var.rate")

sns.histplot(ax = axis[2, 1], data = df, x = "cons.price.idx").set(xlabel = None)
sns.boxplot(ax = axis[3, 1], data = df, x = "cons.price.idx")

sns.histplot(ax = axis[4, 0], data = df, x = "cons.conf.idx").set(xlabel = None)
sns.boxplot(ax = axis[5, 0], data = df, x = "cons.conf.idx")

sns.histplot(ax = axis[4, 1], data = df, x = "euribor3m").set(xlabel = None)
sns.boxplot(ax = axis[5, 1], data = df, x = "euribor3m")

# Ajustar el layout
plt.tight_layout()

# Mostrar el plot
plt.show()

```

```python
# se puede ver que hya una alta concentración entre 30 - 40 años
# los contactos a los clientes en la campaña son cercanos a 1. Hay que revisar los outliers si ese seguimiento de resultado
# hay un gran porcentaje de los clientes que tienen una tasa de empleo positiva. Esto quiere decir que hay un crecimiento y más empleo. + empleo hay más plata para invertir.
# cons.price.idx. Índice de precios al consumidor. Indicador mensual. hay una disminución de un 4-5% en los precio
#cons.conf.idx. Índice de confianza del consumidor. Indicador mensual. Hay una percepción negativa o pesimista sobre la situación economica 
#euribor3m. Tasa EURIBOR 3 meses. Indicador diario (numérico) cercana al 5%
```

```python
#Analisis Categorico-Categorico
# Variables Categoricas: job, marital, education, default, housing, loan, poutcome
# TARGET : y

fig, axis = plt.subplots(4, 2, figsize = (10, 15))

sns.countplot(ax = axis[0, 0], data = df, x = "job", hue = "y").tick_params(axis='x', rotation=90)
sns.countplot(ax = axis[0, 1], data = df, x = "marital", hue = "y").set(ylabel = None)
sns.countplot(ax = axis[1, 0], data = df, x = "education", hue = "y").tick_params(axis='x', rotation=90)
sns.countplot(ax = axis[1, 1], data = df, x = "default", hue = "y")
sns.countplot(ax = axis[2, 0], data = df, x = "housing", hue = "y").set(ylabel = None)
sns.countplot(ax = axis[2, 1], data = df, x = "loan", hue = "y").set(ylabel = None)
sns.countplot(ax = axis[3, 0], data = df, x = "poutcome", hue = "y").set(ylabel = None)

plt.tight_layout()

fig.delaxes(axis[3, 1])

plt.show()

```

```python
# entre mayor la educacion, más posibilidades hay de que tomen un DP
# El hecho de que no tengan un crédito/prestamos asociado, es más posible que tomen un DP
# Vamos a eliminar la variable poutcome, ya que no genera información relevante para tomar una decisición.
```
