```python
# Eliminar información irrelevante
#  identificar a los clientes existentes que tienen una mayor probabilidad de contratar un depósito a largo plazo
# age. SI: la edad si es relevante, ya que es probable que a mayor edad sea más consciente de las inversiones y tengan en mente un deposito a plazo
# relevantes = age, job, marital, education, default, housing, loan, campaign, pdays, poutcome, emp.var.rate, cons.price.idx, cons.conf.idx, EURIBOR 3, y
# irrelevantes = contact, month, day_of_week, duration, previous, nr.employed
```

```python
columnas_eliminar=['contact', 'month', 'day_of_week', 'duration', 'previous', 'nr.employed']
df.drop(columnas_eliminar,axis=1,inplace=True)
df.info()
```

```python
# Analisis univariante
# Analisis Variables categoricas. Una variable categórica es un tipo de variable que puede tomar uno de un número limitado de categorías o grupos
# Hay que usar Histograma e identificar cuales son categoricas
# Variables Categoricas: job, marital, education, default, housing, loan, poutcome, y

import matplotlib.pyplot as plt
import seaborn as sns

fig, axis = plt.subplots(4, 2, figsize = (10, 15))

sns.histplot(ax= axis[0,0], data = df, x='job').tick_params(axis='x', rotation=45)
sns.histplot(ax= axis[0,1], data = df, x='marital').tick_params(axis='x', rotation=45)
sns.histplot(ax= axis[1,0], data = df, x='education').tick_params(axis='x', rotation=90)
sns.histplot(ax= axis[1,1], data = df, x='default').tick_params(axis='x', rotation=45)
sns.histplot(ax= axis[2,0], data = df, x='housing')
sns.histplot(ax= axis[2,1], data = df, x='loan')
sns.histplot(ax= axis[3,0], data = df, x='poutcome')
sns.histplot(ax= axis[3,1], data = df, x='y')

# Ajustar el layout
plt.tight_layout()

plt.show()
```

```python
# hay una presencia mayoritaria de clientes con estudios universitarios
# no tienen cliente vigente 
# un pco más de la mitad del dataset tiene un credito hipotecario, pero una gran participacion no tiene un prestamo personal (credito de consumo)
# El exito de la campaña anterior es muy bajo, ya que gran porcentaje esta identificado sin registro.
# menos de 5 mil han contratado DP
```
