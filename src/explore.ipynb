```python
# Analisis univariante
# Analisis Variables categoricas. Una variable categórica es un tipo de variable que puede tomar uno de un número limitado de categorías o grupos
# Hay que usar Histograma e identificar cuales son categoricas
# Variables Categoricas: job, marital, education, default, housing, loan, poutcome, y

import matplotlib.pyplot as plt
import seaborn as sns

fig, axis = plt.subplots(4, 2, figsize = (10, 15))

sns.histplot(ax= axis[0,0], data = df, x='job').tick_params(axis='x', rotation=45)
sns.histplot(ax= axis[0,1], data = df, x='marital').tick_params(axis='x', rotation=45)
sns.histplot(ax= axis[1,0], data = df, x='education').tick_params(axis='x', rotation=90)
sns.histplot(ax= axis[1,1], data = df, x='default').tick_params(axis='x', rotation=45)
sns.histplot(ax= axis[2,0], data = df, x='housing')
sns.histplot(ax= axis[2,1], data = df, x='loan')
sns.histplot(ax= axis[3,0], data = df, x='poutcome')
sns.histplot(ax= axis[3,1], data = df, x='y')

# Ajustar el layout
plt.tight_layout()

plt.show()
```

```python
#Analisis Categorico-Categorico
# Variables Categoricas: job, marital, education, default, housing, loan, poutcome
# TARGET : y

fig, axis = plt.subplots(4, 2, figsize = (10, 15))

sns.countplot(ax = axis[0, 0], data = df, x = "job", hue = "y").tick_params(axis='x', rotation=90)
sns.countplot(ax = axis[0, 1], data = df, x = "marital", hue = "y").set(ylabel = None)
sns.countplot(ax = axis[1, 0], data = df, x = "education", hue = "y").tick_params(axis='x', rotation=90)
sns.countplot(ax = axis[1, 1], data = df, x = "default", hue = "y")
sns.countplot(ax = axis[2, 0], data = df, x = "housing", hue = "y").set(ylabel = None)
sns.countplot(ax = axis[2, 1], data = df, x = "loan", hue = "y").set(ylabel = None)
sns.countplot(ax = axis[3, 0], data = df, x = "poutcome", hue = "y").set(ylabel = None)

plt.tight_layout()

fig.delaxes(axis[3, 1])

plt.show()

```
